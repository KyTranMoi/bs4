```python
def compare_encodings(example_string: str) -> None:
    encoding = tiktoken.get_encoding("gpt2")
    token_integers = encoding.encode(example_string)
    num_tokens = len(token_integers)
    return num_tokens
        
```

```python
import os
import json

# Đường dẫn đến thư mục chứa các file JSON
folder_path = f'D:\CodeNguon\selenium\\box_chat\code_crawl\data'

# Tạo một từ điển để lưu trữ các tags và nội dung
tag_content_dict = {}

# Duyệt qua các file trong thư mục
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):  # Chỉ xử lý các file có đuôi .json
        file_path = os.path.join(folder_path, filename)

        # Đọc file JSON
        with open(file_path, encoding='utf-8-sig') as json_file:
            data = json.load(json_file)

        # Xử lý từng đối tượng trong danh sách JSON
        for item in data:
            tags = item.get('tags')
            content = item.get('content')

            if tags and content:
                if isinstance(tags, list):
                    tags = ', '.join(tags)

                if tags in tag_content_dict:
                    tag_content_dict[tags].append(content)
                else:
                    tag_content_dict[tags] = [content]

# Tạo chuỗi từ các tags giống nhau và nội dung tương ứng
result_string = ''
for tags, contents in tag_content_dict.items():
    content_tokens = []
    for content in contents:
        content_tokens.append(str((content)))  # Chuyển đổi số nguyên thành chuỗi
    result_string += f'Tags: {tags}\nTokens: {compare_encodings(" ".join(content_tokens))}\n\n'

# In ra chuỗi kết quả
print(result_string)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sắp xếp DataFrame theo giá trị Tokens giảm dần
df_sorted = df.sort_values(by='Tokens', ascending=False)

# Lấy top 10 tag
top_10_tags = df_sorted.head(9)

# Tính tổng số lượng Tokens cho top 10 tags
top_10_tokens_sum = top_10_tags['Tokens'].sum()

# Tính tỉ lệ phần trăm cho top 10 tags
top_10_tags['Percentage'] = (top_10_tags['Tokens'] / top_10_tokens_sum) * 100

# Tính tỉ lệ phần trăm cho các tag còn lại
other_tags_tokens_sum = df_sorted.iloc[10:]['Tokens'].sum()
other_tags_percentage = (other_tags_tokens_sum / top_10_tokens_sum) * 100

# Thêm tag 'Others' với tỉ lệ phần trăm cho các tag còn lại
top_10_tags = top_10_tags.append({'Tags': 'Còn lại', 'Tokens': other_tags_tokens_sum, 'Percentage': other_tags_percentage}, ignore_index=True)

# Tạo danh sách các nhãn và giá trị tương ứng
labels = top_10_tags['Tags']
sizes = top_10_tags['Percentage']
colors = sns.color_palette('pastel', len(labels))


# Vẽ biểu đồ hình tròn với màu sắc riêng biệt
# colors = sns.color_palette('pastel')[0:len(labels)]
plt.figure(figsize=(10, 10))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=180, textprops={'fontsize': 14})

# Tùy chỉnh hình tròn để có hình dạng tròn
plt.axis('equal')

# Hiển thị biểu đồ
plt.savefig('top_10_tags.png')
plt.show()
```
